```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
```

```python
omc_data = pd.read_csv('/home/mike/Documents/Coding/Data/OMC_Performance.csv')

data = omc_data.copy()

```

```python

data = data.dropna(how='any', axis=0)
data['Date_sent'] = pd.to_datetime(data['Date_sent'])
display(data)
```

```python

barchart = px.bar(
    data_frame=data,
    x='Name_Date_sent',
    y=['Number_of_sent', 'Number_of_opens'],
    opacity=0.9,
    orientation="v",
    barmode='overlay',				# 'overlay' / 'relative' / group
    #labels={"ЗП (руб)":"Желаемая зарплата (руб)",
    #"Name_Date_sent":"OMC - Date"},				# можно прописать названия
    title='Sent & Opens',
    width=1400,
    height=720,
    template='none'
)

barchart.show('png')
```

```python

hist = px.histogram(
    data_frame=data,
    x='Open_Rate',

    title='Open Rate Performance',
    histnorm='percent',
    width=1400,
    marginal='box'
)

hist.show('png')
```

```python
scatter_data = data.sort_values(by=['Number_of_sent', 'Number_of_opens'])

fig = px.scatter(
    scatter_data,
    x='Number_of_sent',
    y='Open_Rate',
    title='Number of sent and opens'
)

fig.show('png')

```
