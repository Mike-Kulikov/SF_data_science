```python
import plotly
import plotly.express as px
```

```python
def update_function(x):
    return 5 if x == 0 else 50
df["Renew"] = df["Renew"].apply(update_function)
```

```python

fig = px.scatter(
    df,
    x='Unique views on CRIISP',
    y='Investors on CRIISP',
    color='Intercom Sessions',
    color_continuous_scale='Turbo',
    size='Renew',
    width=1200,
    height=600,
    title='Correlation between Unique views and Investors on CRIISP + Intercom sessions'
)
fig.show('png')
```

```python

fig = px.scatter(
    df,
    x='Pardot clicks',
    y='Pardot activities',
    color='CSM',
    size='Renew',
    width=1200,
    height=600,
    title='Correlation between the number of clicks and activities on Pardot'
)
fig.show('png')
```
